```python
import matplotlib.pyplot as plt

# 1. 데이터 정의
# "Less than or equal to" 값들 (X축: 상한값)
audit_times = ['<= 14', '<= 19', '<= 24', '<= 29', '<= 34']

# 누적 도수 (Y축)
cumulative_frequency = [4, 12, 17, 19, 20]

# 2. 그래프 그리기 (Ogive Curve)
plt.figure(figsize=(10, 6))

# 선 그래프와 마커 표시 (Ogive는 꺾은선 그래프 형태)
plt.plot(audit_times, cumulative_frequency,
         marker='o',         # 원형 마커
         linestyle='-',      # 실선
         color='blue',       # 선 색상
         linewidth=2,        # 선 굵기
         markersize=8)       # 마커 크기

# 3. 각 포인트에 데이터 값 표시 (Annotation)
for x, y in zip(audit_times, cumulative_frequency):
    plt.text(x, y + 0.5,     # 텍스트 위치 (약간 위쪽)
             str(y),         # 표시할 텍스트
             ha='center',    # 가로 정렬
             fontsize=12,
             fontweight='bold')

# 4. 그래프 꾸미기
plt.title('Cumulative Frequency Distribution: Sanderson and Clifford', fontsize=16)
plt.xlabel('Audit Time (Days)', fontsize=14)
plt.ylabel('Cumulative Frequency', fontsize=14)
plt.xticks(audit_times)      # X축 눈금을 데이터 값에 맞춤
plt.ylim(0, 22)              # Y축 범위 설정 (여유 있게)
plt.grid(True, linestyle='--', alpha=0.7) # 격자 추가

# 5. 출력
plt.show()
```

# 1. Data Preparation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Create a sample dataset
data = {
    'Department': ['Sales', 'Marketing', 'Sales', 'IT', 'Sales', 'Marketing', 'IT', 'Sales', 'Marketing', 'IT'],
    'Years_Experience': [1, 3, 2, 5, 10, 2, 8, 4, 6, 3],
    'Monthly_Sales': [40, 60, 45, 80, 120, 50, 90, 55, 75, 65],
    'Performance_Weight': [1.0, 1.2, 1.0, 1.5, 2.0, 1.2, 1.5, 1.0, 1.2, 1.5]
}
df = pd.DataFrame(data)
display(df)
```

# 2. Measures of Location & Variability

```python
# 1. Measures of Location
mean_sales = df['Monthly_Sales'].mean()
median_sales = df['Monthly_Sales'].median()
mode_dept = df['Department'].mode()[0]
weighted_mean = np.average(df['Monthly_Sales'], weights=df['Performance_Weight'])

# 2. Percentile & Quartiles
q1 = df['Monthly_Sales'].quantile(0.25)
q3 = df['Monthly_Sales'].quantile(0.75)
percentile_90 = df['Monthly_Sales'].quantile(0.90)

# 3. Measures of Variability
sales_range = df['Monthly_Sales'].max() - df['Monthly_Sales'].min()
iqr = q3 - q1

# 4. Variance & Standard Deviation (Sample: ddof=1, Population: ddof=0)
sample_var = df['Monthly_Sales'].var(ddof=1)
pop_var = df['Monthly_Sales'].var(ddof=0)
sample_std = df['Monthly_Sales'].std(ddof=1)
pop_std = df['Monthly_Sales'].std(ddof=0)

# Print results
print("--- Location ---")
print(f"Mean: {mean_sales}, Median: {median_sales}, Mode: {mode_dept}")
print(f"Weighted Mean: {weighted_mean:.2f}")
print(f"Q1: {q1}, Q3: {q3}, 90th Percentile: {percentile_90}")
print("\n--- Variability ---")
print(f"Range: {sales_range}, IQR: {iqr}")
print(f"Sample Variance: {sample_var:.2f}, Population Variance: {pop_var:.2f}")
print(f"Sample Std Dev: {sample_std:.2f}, Population Std Dev: {pop_std:.2f}")
```

# 3. Covariance & Correlation

```python
# Calculate Covariance and Correlation between Experience and Sales
covariance = df['Years_Experience'].cov(df['Monthly_Sales'])
correlation = df['Years_Experience'].corr(df['Monthly_Sales'])

print(f"Covariance: {covariance:.2f}")
print(f"Correlation: {correlation:.2f}")
```

# 4. Frequency & Cumulative Frequency

```python
# 1. Categorical Frequency
freq_category = df['Department'].value_counts()

# 2. Continuous Frequency (Binning 'Monthly_Sales')
bins = [30, 60, 90, 120]
freq_continuous = pd.cut(df['Monthly_Sales'], bins=bins).value_counts().sort_index()

# 3. Cumulative Frequency
cum_freq = freq_continuous.cumsum()

print("--- Categorical Frequency ---")
print(freq_category, "\n")
print("--- Continuous Frequency ---")
print(freq_continuous, "\n")
print("--- Cumulative Frequency ---")
print(cum_freq)
```

# 5. Visualizations

```python
# Set up a 2x2 grid for subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 1. Histogram (Continuous Data)
sns.histplot(df['Monthly_Sales'], bins=4, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Histogram of Monthly Sales')

# 2. Bar Chart (Categorical Data)
df['Department'].value_counts().plot(kind='bar', ax=axes[0, 1], color='lightgreen')
axes[0, 1].set_title('Bar Chart of Departments')
axes[0, 1].tick_params(axis='x', rotation=0)

# 3. Skewness Check (Density Plot)
sns.kdeplot(df['Monthly_Sales'], ax=axes[1, 0], color='salmon', fill=True)
axes[1, 0].set_title('Density Plot (Visualizing Skewness)')

# 4. Boxplot
sns.boxplot(y=df['Monthly_Sales'], ax=axes[1, 1], color='gold')
axes[1, 1].set_title('Boxplot of Monthly Sales')

plt.tight_layout()
plt.show()
```
